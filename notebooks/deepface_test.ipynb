```python
from deepface import DeepFace
import os

db_path = "E:\PORTAFOLIO\Documentos\Proyecto\DeepSec\DeepSecurity\Test_DB"

img_path = "E:\PORTAFOLIO\Documentos\Proyecto\DeepSec\DeepSecurity\Test_DB\Messi\messi_1.jpg"
```

```python
models = [
    "VGG-Face", "Facenet", "Facenet512", "OpenFace", "DeepFace",
    "DeepID", "ArcFace", "Dlib", "SFace", "GhostFaceNet",
    "Buffalo_L",
]

metrics = ["cosine", "euclidean", "euclidean_l2", "angular"]
```

```python
dfs = DeepFace.find(
  img_path = img_path, db_path = db_path, model_name = models[9], distance_metric=metrics[0], threshold=0.9, detector_backend="opencv"
)
```

```python
identity_path = dfs[0].iloc[0]["identity"]
identity_name = os.path.basename(os.path.dirname(identity_path))
identity_name
```
